```python
'''
計算現金股利
使用yfinance呼叫該股Ticker類，並使用info中具備殖利率信息，拿來使用後，
再取dividendYield即可獲得現金股利資訊
'''
#import需要的套件
import yfinance as yf
import pandas as pd
import numpy as np
import time
import datetime
import traceback
import utility_f as uf

#讀取台股列表
stock_list = pd.read_excel('D:/13_Pyhon_金融市場/Trading/stock_list.xlsx')
#獲取所有的股票代號
all_stock = stock_list['代號'].values
#儲存每支股票的殖利率
dividend_store = []
stock_store = []
#計數用
count = 0
#迴圈loop每一支股票
for i in all_stock:
    try:
        #我們來計算一下每一筆處理的時間，在開頭記錄一個start時間點
        start = time.time()
        #計數用參數
        count+=1
        #yfinance呼叫該股Ticker類
        stock = yf.Ticker(f'{i}.TW')
        #info中具備殖利率信息，拿來使用
        d_y = stock.info['dividendYield']
        #有時候yfinance回傳的資料有異常需做排除，如果值不是0才儲存
        if d_y!=None:
            dividend_store.append(d_y)
        #記錄每一筆結束時間
        end = time.time()
        #將進度print出來
        print(f'Dealing: {count} | All: {len(all_stock)} | Stock: {i} | DY: {d_y} | Cost Time: {end-start}s')
    except:
        print(f'Error Stock ! Dealing: {count} | All: {len(all_stock)} | Stock: {i}')
                
```

```python
dividend_info = np.array(dividend_store)

#平均數
print(f'Mean:{np.mean(dividend_store)}')

#最大
print(f'Max:{np.max(dividend_store)}')

#最小
print(f'Min:{np.min(dividend_store)}')

#標準差
print(f'Std:{np.std(dividend_store)}')

#第50分位數
print(f'50%:{np.percentile(dividend_store,50)}')

#第70分位數
print(f'70%:{np.percentile(dividend_store,70)}')

#第90分位數
print(f'90%:{np.percentile(dividend_store,90)}')
```
